```python
import pandas as pd
from pathlib import Path

raw = Path("../data_raw/mock_prices.csv")
out = Path("../data_clean/clean_housing.csv")

df = pd.read_csv(raw)

# Params (simple assumptions for now)
ASSUMED_FLAT_SIZE_M2 = 70
ASSUMED_MORTGAGE_RATE = 0.05  # 5% annual carrying-cost proxy

# Derived indicators
df["price_to_income_ratio"] = (df["price_per_m2"] * ASSUMED_FLAT_SIZE_M2) / (df["avg_monthly_wage"] * 12)
df["rent_vs_buy_index"] = (df["avg_monthly_rent"] * 12) / (df["price_per_m2"] * ASSUMED_FLAT_SIZE_M2 * ASSUMED_MORTGAGE_RATE)

out.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(out, index=False)
print(" Saved", out)
df.head()
```
