```python
import pandas as pd
from pathlib import Path

inp = Path("../data_clean/clean_housing.csv")
out = Path("../data_clean/kpis.csv")

df = pd.read_csv(inp).sort_values(["region","year"])
df["yoy_price_pct"] = df.groupby("region")["price_per_m2"].pct_change() * 100

kpi = (df.groupby(["year","region"], as_index=False)
       .agg(
         avg_price_m2=("price_per_m2","mean"),
         price_to_income=("price_to_income_ratio","mean"),
         rent_buy=("rent_vs_buy_index","mean"),
         avg_wage=("avg_monthly_wage","mean"),
         yoy_price_pct=("yoy_price_pct","mean")
       ))

kpi.to_csv(out, index=False)
print("Saved", out)
kpi
```
